Consideremos otro caso, pero ahora fabriquemos un conjunto de datos conveniente a partir de una generación de números aleatorios

```python
import numpy as np
```

```python
np.random.seed(1995)
a = np.random.multivariate_normal([0,0],[[3,4],[4,3]], size = 100)
b = np.random.multivariate_normal([10,15],[[3,4],[4,3]], size = 50)
X = np.concatenate((a,b))
print(X.shape)
plt.scatter(X[:,0],X[:,1])
plt.grid()
plt.show()
```

Generamos la matriz de enlace

```python
z = linkage(X,"ward")
pd.DataFrame(z).head(20)
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
```

```python
c, coph_dist = cophenet(z,pdist(X))
```

```python
c # coeficiente de correlación cophenetico, compara las distancias originales con respecto a los clustering que se
# han generado
```

```python
#notemos que el punto correspondiente al índice 82 es el primer en juntarse con un cluster ya establecido

z[154-len(X)]

idx = [38,92,82] # Recordemos que cada indice representa un par de puntos en el plano

plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1])
plt.scatter(X[idx,0],X[idx,1], c="r")
plt.grid()
plt.show()
```

**Dendrograma**

```python
fig = plt.figure(figsize = (25,10))
plt.title("Dendrograma del Clustering Jerárquico")
plt.xlabel("Índices de la muestra")
plt.ylabel("Distancias")
dendrogram(z,leaf_rotation=90, leaf_font_size=8, color_threshold=0.7*154)
plt.show()
```

**Truncando el dendrograma**

```python
fig = plt.figure(figsize = (25,10))
plt.title("Dendrograma del Clustering Jerárquico truncado")
plt.xlabel("Índices de la muestra")
plt.ylabel("Distancias")
dendrogram(z,leaf_rotation=90, leaf_font_size=15, color_threshold=0.7*154, truncate_mode = "lastp",
          p = 10, show_leaf_counts=True, show_contracted=True)
plt.show()
```

```python
def dendrogram_tune(*args, **kwargs):  # Explicación sobre *args, **kwargs, compañero dejo link en preguntas del video
    
    max_d=kwargs.pop("max_d", None) 
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)
    
    plt.figure(figsize=(16,8))
    ddata = dendrogram(*args,**kwargs)
        
    if not kwargs.get('no_plot', False):
        plt.title("Clustering jerárquico con Dendrograma truncado")
        plt.xlabel("Índice del Dataset (o tamaño del cluster)")
        plt.ylabel("Distancia")
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y>annotate_above:
                plt.plot(x,y,'o',c=c)
                plt.annotate('%.3g'%y, (x,y), xytext=(0,-5),
                            textcoords="offset points", va="top", ha="center")
                
    if max_d:
        plt.axhline(y=max_d, c='k')
        
    return ddata
```

```python
dendrogram_tune(z,truncate_mode='lastp',p=12, leaf_rotation=90., leaf_font_size=12.,
                show_contracted=True,annotate_above=10, max_d=20) # a altura 20 se formarian 4 clusters
plt.show()
```

**Metodo del codo**

Para encontrar el número optimo de clusters:

```python
last = z[-10:,2]
last_rev = last[::-1]
idx = np.arange(1,len(last)+1)
plt.plot(idx,last_rev)

acc = np.diff(last,2) #Aceleración, buscamos la máxima aceleración
acc_rev = acc[::-1]
plt.plot(idx[:-2]+1,acc_rev)
plt.show()
k = acc_rev.argmax()+2
print("El número optimo de cluster es %s"%str(k))
```
