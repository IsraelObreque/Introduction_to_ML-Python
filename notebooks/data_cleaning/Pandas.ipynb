# Carga de datos a través de la función open

Este método sera muy utilizado para datasets muy grandes, para estos casos sera mejor cargar los datos de manera "manual", es decir, usando un ciclo for para recorrer el fichero linea por linea.

Haremos una carga de datos mediante la función open y expresaremos nuestro carga mediante un dataframe para así obtener el mismo resultado que obtenemos mediante el módulo de pandas


```python
data3 = open(mainpath + "/customer-churn-model/Customer Churn Model.txt")
cols = data3.readline().strip().split(',')
main_dict = {}
n_cols = len(cols)
```

```python
for col in cols:
    main_dict[col] = []
```

```python
counter  = 0
for line in data3:
    values = line.strip().split(',')
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter +=1
print("El archivo tiene %d filas y %d columnas"%(counter,n_cols))
```

```python
df = pd.DataFrame(main_dict)
```

```python
df.head()
```

Podemos resumir lo hecho anteriormente en una función

```python
def carga_manual(path):
    data = open(path)
    cols = data.readline().strip().split(',')
    main_dict = {}
    n_cols = len(cols)

    for col in cols:
        main_dict[col] = []

    counter  = 0
    for line in data:
        values = line.strip().split(',')
        for i in range(n_cols):
            main_dict[cols[i]].append(values[i])
        counter +=1
    print("El archivo tiene %d filas y %d columnas"%(counter,n_cols))
    
    df = pd.DataFrame(main_dict)
    
    return df
```

```python
data4 = carga_manual(mainpath + "/customer-churn-model/Customer Churn Model.txt")
```

*nota:* esta es una función bastante básica, de hecho, ya tendría problemas con *titanic3.csv* debido a que hay columnas como *name* que tienen strings con comas de por medio como valores.

# Lectura y escritura de ficheros

En este cas, tomaremos un fichero conocido y cambiaremos el separador de '\t' a ','

```python
infile = mainpath + '/customer-churn-model/Tab Customer Churn Model.txt'
outfile = mainpath + '/customer-churn-model/customer churn model.txt'

with open(infile,'r') as infile1:
    with open(outfile,'w') as outfile1:
        for line in infile1:
            fields = line.strip().split('\t')
            outfile1.write(','.join(fields))
            outfile1.write('\n')
```

```python
df4 = pd.read_csv(outfile)
```

Pandas tiene una función para realizar la misma tarea, en este caso transformaremos nuestro DataFrame a formato xls

```python
df4.to_excel(mainpath+'/customer-churn-model/customer.xls')
```

# Leer datos desde una URL

```python
medals_url = "http://winterolympicsmedals.com/medals.csv"
medals_data = pd.read_csv(medals_url)
medals_data.head()
```

**Ejercicio de descarga de datos con urllib3**

Veamos un ejemplo usando la libreria urllib3 para leer datos desde una URL externa, procesarlos y convertirlos en un DataFrame de python antes de guardarlo en una CSV local.

```python
import urllib3
http = urllib3.PoolManager()
r = http.request('GET', medals_url)
print("El estado de la respuesta es: ",r.status)
response = r.data

str_data = response.decode("utf-8")

lines = str_data.split("\n")

col_names = lines[0].split(',')

main_dict= {}

for col in col_names:
    main_dict[col] = []
    
for line in lines:
    values = line.strip().split(',')
    for i in range(len(col_names)):
        main_dict[col_names[i]].append(values[i])

data = pd.DataFrame(main_dict)

mainpath = "/home/israel/Documentos/Machine-learning/python-ml-course/datasets"
filepath="athletes/downloaded_medals"
fullpath=os.path.join(mainpath,filepath)

data.to_csv(fullpath+'.csv')
data.to_json(fullpath+'.json')
data.to_excel(fullpath+'.xls')

data.head()
```

Podemos guardar esto en una función

```python
def carga_url(url,path):
    
    import urllib3
    http = urllib3.PoolManager()
    r = http.request('GET', url)
    print("El estado de la respuesta es: ",r.status)
    response = r.data

    str_data = response.decode("utf-8")

    lines = str_data.split("\n")

    col_names = lines[0].split(',')

    main_dict= {}

    for col in col_names:
        main_dict[col] = []

    for line in lines:
        values = line.strip().split(',')
        for i in range(len(col_names)):
            main_dict[col_names[i]].append(values[i])

    data = pd.DataFrame(main_dict)

    data.to_csv(path+'.csv')
    data.to_json(path+'.json')
    data.to_excel(path+'.xls')

    return data
```

```python
data =  carga_url(medals_url, fullpath)
```
